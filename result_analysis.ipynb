```python
fig, axs = plt.subplots(4, 2, figsize=(10, 18))
#fig.suptitle('Different context for single channel / single + sum')
fig.tight_layout()
confusion_matrix_path = './result/confusion_matrix/'
for i in range(8):
    exp = exps[i]
    experiment_name, epoch = exp
    dataset = 'test' # test of train
    file_path = confusion_matrix_path+experiment_name+'_epoch'+epoch+'_'+dataset+'.npy'
    con_mat = np.load(file_path)
    print('con_mat ', i, ' shape = ', con_mat.shape)
    print('pos = ', math.floor(i/2), ', ', i%2)
    sns.heatmap(con_mat, annot=True, ax=axs[i%4][math.floor(i/4)], fmt='.2f')

    ticks = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5]
    labels = ['None', 'mel', 'rhy', 'mel+rhy', 'harm', 'mel+harm', 'rhy+harm', 'mel+rhy+harm']
    axs[i%4][math.floor(i/4)].set_xticks(ticks, labels, fontsize=8, rotation=45)
    axs[i%4][math.floor(i/4)].set_yticks(ticks, labels, fontsize=8, rotation=45)
    axs[i%4][math.floor(i/4)].set_title(titles[i])
    plt.subplots_adjust(wspace=0.2, hspace=0.3)

save_path = './result/analysis/'+'conmats_different_channels_with_context=0.jpg'
plt.savefig(save_path)
print('figure saved at ', save_path)
```
